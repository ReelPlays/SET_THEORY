#4 $A \cup B = A \cup (B \cap A^c)$

To prove the set identity $A \cup B = A \cup (B \cap A^c)$, we'll demonstrate that both sets are equal by showing that any element in one set is also in the other set.

Let's break it down:

Definition:
- $A \cup B$ represents the union of sets $A$ and $B$, containing elements that are in either $A$ or $B$.
- $B \cap A^c$ represents elements in $B$ that are not in $A$ (complement of $A$ intersected with $B$).

Proof:

1. $A \cup B \subseteq A \cup (B \cap A^c)$:

   Take an element $x$ in $A \cup B$. This means $x$ is either in $A$ or in $B$ (or in both). Therefore, $x$ is also in $A \cup (B \cap A^c)$ because it's in $A$ or in the intersection of $B$ and $A^c$.

2. $A \cup (B \cap A^c) \subseteq A \cup B$:

   Take an element $y$ in $A \cup (B \cap A^c)$. This means $y$ is either in $A$ or in the intersection of $B$ and $A^c$. In either case, $y$ is in $A \cup B$ because it's in $A$ or in $B$ (or in both).

Since any element in $A \cup B$ is also in $A \cup (B \cap A^c)$, and vice versa, we conclude that $A \cup B = A \cup (B \cap A^c)$.

To prove the Inclusion-Exclusion Principle:

Let $A$ and $B$ be two sets. We want to find the cardinality of their union $A \cup B$, denoted as $|A \cup B|$. We'll demonstrate that $|A \cup B| = |A| + |B| - |A \cap B|$.

We'll break it down into three cases:

1. Elements that are in set $A$ but not in set $B$: $|A \setminus B| = |A| - |A \cap B|$.

2. Elements that are in set $B$ but not in set $A$: $|B \setminus A| = |B| - |A \cap B|$.

3. Elements that are in both sets $A$ and $B$: $|A \cap B|$.

Now, we'll calculate the cardinality of the union:

\begin{align}
|A \cup B| &= |(A \setminus B) \cup (B \setminus A) \cup (A \cap B)| \nonumber \\
&= |A \setminus B| + |B \setminus A| + |A \cap B| \nonumber \\
&= (|A| - |A \cap B|) + (|B| - |A \cap B|) + |A \cap B| \nonumber \\
&= |A| + |B| - |A \cap B| \nonumber
\end{align}

So, we have shown that $|A \cup B| = |A| + |B| - |A \cap B|$, which is the Inclusion-Exclusion Principle.
