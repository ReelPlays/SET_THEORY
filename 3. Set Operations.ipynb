```python
#1. C XOR D

from matplotlib import pyplot as plt
from matplotlib_venn import venn2

# Define the sets C and D
C = {5, 6, 7, 8, 9}
D = {1, 3, 5, 7, 9}

# Calculate XOR (Symmetric Difference)
xor_result = C.symmetric_difference(D)

# Create a Venn diagram for XOR
venn = venn2([C, D], ('C', 'D'))
venn.get_label_by_id('100').set_text('C - D')
venn.get_label_by_id('010').set_text('D - C')
venn.get_label_by_id('110').set_text('C âˆ© D')

plt.title("Venn Diagram for C XOR D")
plt.show()

print("C XOR D:", xor_result)
```

```python
#2. A\B

from matplotlib import pyplot as plt
from matplotlib_venn import venn2

# Define the universal set U and sets A, B
U = {1, 2, 3, 4, 5}
A = {1, 2, 3, 4, 5}
B = {4, 5, 6, 7}

# Calculate A \ B (Set difference)
difference_result = A.difference(B)

# Create a Venn diagram for A \ B
venn = venn2([A, B], ('A', 'B'))
venn.get_label_by_id('100').set_text('A - B')
venn.get_label_by_id('010').set_text('B')
venn.get_label_by_id('110').set_text('A âˆ© B')

plt.title("Venn Diagram for A \\ B")
plt.show()

print("A \\ B:", difference_result)






```
