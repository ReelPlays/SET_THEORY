```python
#1
from matplotlib import pyplot as plt
from matplotlib_venn import venn3

# Define the sizes of each set
set_sizes = (65, 45, 42, 20, 25, 15, 8)

# Create a Venn diagram
venn = venn3(subsets=set_sizes, set_labels=('Newsweek', 'Time', 'Fortune'))

# Calculate the number of people who read at least one magazine
at_least_one = sum(set_sizes) - int(venn.get_label_by_id('100').get_text().replace('\n', ''))
print("Number of people who read at least one magazine:", at_least_one)

# Find the number of people who read exactly one magazine
exactly_one = (
    int(venn.get_label_by_id('100').get_text()) +
    int(venn.get_label_by_id('010').get_text()) +
    int(venn.get_label_by_id('001').get_text())
)
print("Number of people who read exactly one magazine:", exactly_one)

plt.title("Venn Diagram of Magazine Readers")
plt.show()
```

```python
#2
from matplotlib import pyplot as plt
from matplotlib_venn import venn3

# Define the sizes of each set
set_sizes = (15, 12, 11, 3, 9, 4, 5)

# Create a Venn diagram
venn = venn3(subsets=set_sizes, set_labels=('A', 'R', 'W'))

# Find the number of cars with W only
W_only = set_sizes[2] - int(venn.get_label_by_id('010').get_text().split("\n")[0])
print("Number of cars with W only:", W_only)

# Find the number of cars with A only
A_only = set_sizes[0] - int(venn.get_label_by_id('100').get_text().split("\n")[0])
print("Number of cars with A only:", A_only)

# Find the number of cars with R only
R_only = set_sizes[1] - int(venn.get_label_by_id('001').get_text().split("\n")[0])
print("Number of cars with R only:", R_only)

# Find the number of cars with both R and W but not A
R_and_W_not_A = int(venn.get_label_by_id('001').get_text().split("\n")[0])
print("Number of cars with both R and W but not A:", R_and_W_not_A)

# Find the number of cars with both A and R but not W
A_and_R_not_W = int(venn.get_label_by_id('100').get_text().split("\n")[0])
print("Number of cars with both A and R but not W:", A_and_R_not_W)

# Find the number of cars with only one of the options
only_one_option = A_only + R_only + W_only
print("Number of cars with only one of the options:", only_one_option)

# Find the number of cars with at least one option
at_least_one_option = sum(set_sizes) - int(venn.get_label_by_id('100').get_text().split("\n")[0]) - int(venn.get_label_by_id('010').get_text().split("\n")[0]) - int(venn.get_label_by_id('001').get_text().split("\n")[0])
print("Number of cars with at least one option:", at_least_one_option)

# Find the number of cars with none of the options
none_of_the_options = 25 - at_least_one_option
print("Number of cars with none of the options:", none_of_the_options)

plt.title("Venn Diagram of Car Options")
plt.show()
```
